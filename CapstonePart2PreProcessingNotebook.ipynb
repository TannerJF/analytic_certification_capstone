**NULL VALUE ANALYSIS**

```python
print(bank_df.shape)

print(bank_df.isnull().sum())
```

**IMPUTATION CODING**

```python
#with a majority of the cards--9315 precisely--in "Blue" category, this was used for imputation of the null values.
#being categorized as "object" rather than any numeric value, methods of analysis like mean or median were not viable#

print(bank_df["Card_Category"].value_counts())
bank_df.loc[bank_df["Card_Category"].isnull(),"Card_Category"]="Blue"
```

```python
#With a numeric value like "Credit_Limit" missing, for this data set mean was used#
#if this were an actual real-world situation, the bank would need to be consulted about the factors surrounding the limit#
#arbitrarily alloting a credit limit seems irrisponsible at best#

bank_df.loc[bank_df["Credit_Limit"].isnull(),"Credit_Limit"]=round(bank_df["Credit_Limit"].mean())
```
